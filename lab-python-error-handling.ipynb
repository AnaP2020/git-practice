````python
products = ["t-shirt", "mug", "hat", "book", "keychain"]
````

````python
def initialize_inventory(products):
      inventory = {}
      for product in products:
          while True:
              try:
                  quantity = int(input(f"Enter the quantity of {product}s available: "))
                  if quantity >= 0:
                      inventory[product] = quantity
                      break
                  else:
                      print("Quantity cannot be negative. Please enter a non-negative number.")
              except ValueError:
                  print("Invalid input. Please enter a valid number for the quantity.")
      formatted_items = [f"Enter the quantity of {product}s available: {quantity}" for product, quantity in inventory.items()]
      for item in formatted_items:
        print(item)
      return inventory
````

````python
def get_customer_orders(inventory):
    num_orders = []
    while True:
        try:
            num_orders_input = int(input("Enter the number of products you wish to order: "))
            if num_orders_input >= 0:
                num_orders = num_orders_input
                break
            else:
                print("Please enter a valide number.")
        except ValueError:
            print("Invalid input. Please enter a number for the number of products.")
    print(f"Enter the number of customer orders: {num_orders}")
    orders = set()
    while len(orders) < num_orders:
        try:
            order = input(f"Enter product name {len(orders)+1}: ")
            if order not in inventory:
                print("The product is not available in the inventory.")
            elif inventory[order] <= 0:
                print("The product is out of stock.")
            else:
                orders.add(order)
        except ValueError:
            print("Invalid input. Please enter a valid product name.")
    formatted_orders = [f"Enter the name of a product that a customer wants to order: {order}" for order in sorted(orders)]
    for order in formatted_orders:
        print(order)
    return orders
````

````python
def calculate_order_total_price(customer_orders):
    prices = []

    while len(prices) < len(customer_orders):
        try:
            for order in customer_orders:
                if len(prices) < len(customer_orders):
                    price = float(input(f"Enter the price of {order}: "))
                    if price <= 0:
                        print("Invalid input. Please enter a valid number price.")
                    else:
                        prices.append(price)
        except ValueError:
            print("Invalid input. Please enter a valid number price.")

    for order, price in zip(customer_orders, prices):
        print(f"The price of {order}: {price}")

    total_price = sum(prices)
    print(f"Total price: {total_price:.2f}")
    
    return total_price
````

````python
inventory = initialize_inventory(products)
````

````python
customer_orders = get_customer_orders(inventory)
````

````python
calculate_order_total_price(customer_orders)
````
